```python
import kagglehub

path = kagglehub.dataset_download("adharshinikumar/screentime-vs-mentalwellness-survey-2025")
```

```python
import os
print(os.listdir(path))
```

```python
import pandas as pd
import os
file_path = os.path.join(path, os.listdir(path)[0])
df = pd.read_csv(file_path)
```

```python
df = df.drop(columns=["Unnamed: 15"])
```

```python
df.head()
```

```python
df.plot.scatter(x ="exercise_minutes_per_week" , y ="stress_level_0_10")
plt.title("Exercise vs Stress")
plt.show()
```

```python
df.plot.scatter(x="sleep_hours" , y="mental_wellness_index_0_100")
plt.title("Sleep Hours vs Mental Wellness")
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x="screen_time_hours", y="stress_level_0_10", data=df)

plt.title("Screen Time vs Stress Level")
plt.show()
```

```python
print(df.columns)
```

```python
# sns.regplot(x="work_screen_hours", y="stress_level_0_10", data=df)
sns.regplot(x="leisure_screen_hours", y="stress_level_0_10", data=df)
```

```python
df["screen_group"] = pd.cut(
    df["screen_time_hours"],
    bins=[0,5,8,12,20],
    labels=["Low","Medium","High","Very High"]
)
```

```python
sns.barplot(x="screen_group", y="stress_level_0_10", data=df)
```

```python
df[["screen_time_hours", "work_screen_hours", "leisure_screen_hours"]].head()
```

```python
df["work_ratio"] = df["work_screen_hours"] / df["screen_time_hours"]
df["leisure_ratio"] = df["leisure_screen_hours"] / df["screen_time_hours"]
```

```python
sns.regplot(x="leisure_ratio", y="mental_wellness_index_0_100", data=df)
```

```python
df.groupby("work_mode")["stress_level_0_10"].mean()
```

```python
sns.regplot(x="work_ratio", y="stress_level_0_10", data=df)
```
